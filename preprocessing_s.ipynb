```python
import pandas as pd
import numpy as np

pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv(r'../pandas_df2.csv', dtype='str')
df.head()
```

```python
#keep only the columns we deem necessary
df2 = df[['Name', 'AT_MaraMatkl', 'AT_MaraMaktx', 'AT_MaraBrgew', 'AT_MaraMtart', 'AT_MaraLabor']]
df2.head()
```

```python
# text preprocessing
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords') 
nltk.download('wordnet')

# convert all text to lowercase
df2 = df2.apply(lambda x: x.astype(str).str.lower())
df3 = df2.head(1000).copy()

df3.head()
```

```python
# Make a list of english stopwords
stopwords = nltk.corpus.stopwords.words("english")
#stopwords.extend('"do not use anymore"')

stopwords_removed_df = pd.DataFrame(columns=(lambda col: col for col in df3.columns))

for col in df3.columns:
    stopwords_removed_df[col] = df3[col].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
stopwords_removed_df.head()
```

```python
nltk.download('omw-1.4')
```

```python
# lemmatization
lemmatizer = WordNetLemmatizer()
for col in df3.columns:
    df3[col] = stopwords_removed_df[col].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))

df3.head()
```

```python
# aggregate columns for heavy text clean
agg_cols_df = df3.astype(str).agg(' '.join, axis=1)
agg_cols_df.head()
```

```python
# remove all non-alphanumeric characters
agg_cols_df.replace(r'[^a-zA-Z]', ' ', regex=True, inplace=True)

# remove single characters labels entries
agg_cols_df = agg_cols_df.astype(str).apply(lambda x: ' '.join([w for w in x.split() if len(w)>1]))

# remove uneccessary spaces
agg_cols_df = agg_cols_df.astype(str).apply(lambda x: ' '.join(x.split()))

# lower case
agg_cols_df = agg_cols_df.astype(str).apply(lambda x: x.lower())

agg_cols_df.head()
```

```python
# remove duplicates per row
no_duplicates_df = agg_cols_df.apply(lambda x: ' '.join(set(x.split())))
no_duplicates_df
```
